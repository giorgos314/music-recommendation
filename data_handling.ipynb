The original data comes from a ~5GB SQLite database. <br>
Source: https://www.kaggle.com/datasets/maltegrosse/8-m-spotify-tracks-genre-audio-features <br>
I transformed all its tables into CSV files in order to import them to Pandas and bring the data to the format I want.


```python
# importing all required modules
import json

import numpy as np
import pandas as pd

from pathlib import Path
```

```python
# columns: id, analysis_url, acousticness, danceability, duration, energy, instrumentalness, key, liveness, loudness, mode, speechiness, tempo, time_signature, valence
features_file = Path(r'G:\Thesis\Data\8mil\audio_features.csv')
features = pd.read_csv(features_file)
features = features.drop('analysis_url', axis=1)
```

```python
# columns: name, id, popularity, followers
artists_file = Path(r'G:\Thesis\Data\8mil\artists.csv')
artists = pd.read_csv(artists_file)
```

```python
# columns: id, duration, explicit, audio_feature_id, name, popularity
tracks_file = Path(r'G:\Thesis\Data\8mil\tracks.csv')
tracks = pd.read_csv(tracks_file)
tracks = tracks.drop(['disc_number', 'preview_url', 'track_number', 'is_playable'], axis=1)
```

```python
# Every track and genre in the other tables have a unique id. This table links all artists to multiple genres.
ag_file = Path(r'G:\Thesis\Data\8mil\r_artist_genre.csv')
ag = pd.read_csv(ag_file)
```

```python
# Every track and artist in the other tables have a unique id. This table links all tracks to an artist.
# columns: track_id, artist_id
at_file = Path(r'G:\Thesis\Data\8mil\r_track_artist.csv')
at = pd.read_csv(at_file)
```

```python
# Each artist is linked to multiple genres. This means that multiple entries in the format "artist: genre" exist per artist.
# This code transforms the table to ensure there is only one entry per artist, in the format "artist: [genres]", where [genres]
# is a list of all the genres linked to the artist.
ag_modified = ag.groupby(ag['artist_id'])
ag_modified = ag_modified.agg(', '.join).reset_index()
ag_modified
```

```python
# merging a-t (artist-track table) with the table that holds all artist info
dataset = at.merge(artists, left_on="artist_id", right_on="id")
dataset = dataset.drop(['artist_id', 'id', 'popularity', 'followers'], axis=1)

# merging the dataset we created above with the table that holds all track info
dataset = dataset.merge(tracks, left_on="track_id", right_on="id")
dataset = dataset.drop(['track_id', 'id', 'explicit'], axis=1)

# merging the dataset with the table that holds all track feature info
dataset = dataset.merge(features, left_on="audio_feature_id", right_on="id")
dataset = dataset.drop(['audio_feature_id', 'id', 'duration_x'], axis=1)
dataset = dataset.rename(columns={"name_x": "artist", "duration_y": "duration", "name_y": "song"})

dataset
```
